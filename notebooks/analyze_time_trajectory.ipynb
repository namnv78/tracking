```python
df = pd.read_csv("/content/drive/MyDrive/Viettel DTalent/ViettelTalentsPhase2/Phan Nhật Minh/track_v2/output/MOT20/mot20_public_log.txt",
                 names=['n_det', 'n_track', 't_det', 't_reid', 't_euclid', 't_traject', 't_cosine', 't_hungarian'])
df['det_track'] = df.apply(lambda row: row['n_det']*row['n_track'], axis=1)
```
